```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import HeteroConv, GCNConv
from sklearn.metrics import (
    roc_auc_score, 
    precision_recall_fscore_support, 
    accuracy_score
)
import numpy as np

######################
# 1) HeteroEdgeClassifier Model
######################
class HeteroEdgeClassifier(nn.Module):
    def __init__(self, hidden_channels, out_channels, num_classes=2):
        """
        A semi-supervised GNN for edge classification (attack vs. benign).
        hidden_channels, out_channels: GNN/MLP dimensionalities
        num_classes=2 => two classes: 0 (benign), 1 (attack)
        """
        super().__init__()
        # 1) GNN for node embeddings - define HeteroConv with separate GCNConv for each edge type
        self.conv = HeteroConv({
            ('ip','network','ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
            ('ip','context','ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
            ('ip','knowledge','ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
        }, aggr='sum')

        # 2) MLP to transform node embeddings after the HeteroConv
        self.lin_node = nn.Sequential(
            nn.Linear(hidden_channels, hidden_channels * 2),
            nn.ReLU(),
            nn.Linear(hidden_channels * 2, out_channels),
            nn.ReLU()
        )

        # 3) Edge classification head: from (z_src||z_dst) -> (2) classes
        self.lin_edge = nn.Sequential(
            nn.Linear(out_channels * 2, out_channels),
            nn.ReLU(),
            nn.Linear(out_channels, num_classes)  # 2 => (benign, attack)
        )

    def forward(self, data):
        """
        data: A HeteroData object containing node/edge info.
        returns x_dict with x_dict['ip'] => node embeddings [num_nodes, out_channels]
        """
        x_dict = self.conv(data.x_dict, data.edge_index_dict)
        x_dict['ip'] = self.lin_node(x_dict['ip'])
        return x_dict

    def edge_classify(self, z_src, z_dst):
        """
        Given source/destination node embeddings, produce logits [num_edges, 2].
        """
        z_cat = torch.cat([z_src, z_dst], dim=-1)  # shape [E, out_channels*2]
        logits = self.lin_edge(z_cat)             # shape [E,2]
        return logits


##############################
# 2) Focal Loss
##############################
def focal_loss(logits, targets, alpha=0.25, gamma=2.0):
    """
    logits: [E,2], targets: [E] in {0,1}
    alpha, gamma: focal loss hyperparams
    """
    ce_loss = F.cross_entropy(logits, targets, reduction='none')
    pt = torch.exp(-ce_loss)
    return (alpha * (1 - pt)**gamma * ce_loss).mean()


##############################
# 3) Train Edge Classifier
##############################
def train_edge_classifier(
    model, 
    graphs, 
    epochs=10, 
    device='cuda', 
    alpha=0.25, 
    gamma=2.0
):
    """
    Semi-supervised training with focal_loss on each edge type that has 'edge_label'.
    """
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-3)

    for epoch in range(1, epochs + 1):
        model.train()
        total_loss = 0.0

        for data in graphs:
            data = data.to(device)
            optimizer.zero_grad()

            # 1) Node embeddings
            x_dict = model(data)
            z_ip = x_dict['ip']

            # 2) Sum classification loss across edge types
            loss_graph = 0.0
            for rel_type in data.edge_types:
                if 'edge_label' not in data[rel_type]:
                    continue
                labels = data[rel_type].edge_label
                if labels.numel() == 0:
                    continue

                edge_index = data[rel_type].edge_index
                src, dst = edge_index
                logits = model.edge_classify(z_ip[src], z_ip[dst])
                
                # Focal loss
                loss_rel = focal_loss(logits, labels, alpha=alpha, gamma=gamma)
                loss_graph += loss_rel

            loss_graph.backward()
            optimizer.step()
            total_loss += loss_graph.item()

        avg_loss = total_loss / len(graphs)
        print(f"[Epoch {epoch}/{epochs}] Loss: {avg_loss:.4f}")


################################
# 4) Evaluate Edge Classifier
################################
@torch.no_grad()
def evaluate_edge_classifier(model, test_graphs, device='cuda'):
    """
    Evaluate on test graphs: gather anomaly scores => compute AUC, prec, recall, F1, accuracy.
    """
    model.eval()
    model.to(device)

    all_scores = []
    all_labels = []

    for data in test_graphs:
        data = data.to(device)
        x_dict = model(data)
        z_ip = x_dict['ip']

        for rel_type in data.edge_types:
            if 'edge_label' not in data[rel_type]:
                continue
            labels = data[rel_type].edge_label
            if labels.numel() == 0:
                continue

            edge_index = data[rel_type].edge_index
            src, dst = edge_index
            logits = model.edge_classify(z_ip[src], z_ip[dst])  # [E,2]

            # Probability of class=1 ("attack")
            probs = F.softmax(logits, dim=-1)[:,1]
            all_scores.extend(probs.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())

    all_scores = np.array(all_scores)
    all_labels = np.array(all_labels)

    if len(np.unique(all_labels)) < 2:
        print("Only one class in test set => AUC=0.5 by definition.")
        return

    auc = roc_auc_score(all_labels, all_scores)
    preds = (all_scores > 0.5).astype(int)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_labels, preds, average='binary', zero_division=0
    )
    accuracy = accuracy_score(all_labels, preds)

    print(f"AUC={auc:.4f}, Precision={precision:.4f}, "
          f"Recall={recall:.4f}, F1={f1:.4f}, Accuracy={accuracy:.4f}")
###############################
# 4) Data-Loading Helpers
###############################
def load_single_graph(directory, filename):
    """
    Load a single .pt file from a directory as a HeteroData object.

    Args:
        directory (str): Path to the directory containing the graph file.
        filename (str): Name of the file to load.

    Returns:
        HeteroData: The loaded graph.
    """
    path = os.path.join(directory, filename)
    if os.path.exists(path):
        return torch.load(path)
    else:
        raise FileNotFoundError(f"Graph file {filename} not found in {directory}.")

###############################
# 5) Data-Loading Helpers
###############################
def load_single_graph(directory, filename):
    """
    Load a .pt (HeteroData).
    Remap any label <0 =>0, >1 =>1 to ensure they're in {0,1}.
    """
    path = os.path.join(directory, filename)
    if not os.path.exists(path):
        raise FileNotFoundError(f"{path} not found.")

    data = torch.load(path)  # If you control the files, keep weights_only=False

    # Force labels in [0,1]
    for rel_type in data.edge_types:
        if 'edge_label' in data[rel_type]:
            lbl = data[rel_type].edge_label
            lbl = torch.where(lbl < 0, torch.tensor(0, device=lbl.device), lbl)
            lbl = torch.where(lbl > 1, torch.tensor(1, device=lbl.device), lbl)
            data[rel_type].edge_label = lbl

    return data


#########################
# 6) Main Script
#########################
if __name__ == "__main__":
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # A) Load Train Graphs
    train_dir = "3ed_trai_h_graphs_hetero_graphs"
    train_files = [f for f in os.listdir(train_dir) if f.endswith('.pt')]
    train_graphs = [load_single_graph(train_dir, 'test_graph_hour_14.pt'),load_single_graph(train_dir, 'test_graph_hour_15.pt'),
                    load_single_graph(train_dir, 'test_graph_hour_16.pt'), load_single_graph(train_dir, 'test_graph_hour_17.pt'),
                    load_single_graph(train_dir, 'test_graph_hour_18.pt'), load_single_graph(train_dir, 'test_graph_hour_19.pt'),
                    load_single_graph(train_dir, 'test_graph_hour_20.pt')]
    #train_graphs = [load_single_graph(train_dir, f) for f in train_files]
    print(f"Loaded {len(train_graphs)} training graphs.")

    # B) Initialize Model
    hidden_channels = 32
    out_channels = 16
    model = HeteroEdgeClassifier(hidden_channels, out_channels)

    # C) Train
    train_edge_classifier(model, train_graphs, epochs=1000, device=device)

    # D) Load Test Graphs - make sure the wrds are correct: what is used during training: train and val, or train an test; what is used in eval? test or val?
    test_dir = "3ed_tes_h_graphs_hetero_graphs"
    test_files = [f for f in os.listdir(test_dir) if f.endswith('.pt')]
    test_graphs = [load_single_graph(test_dir, f) for f in test_files]
    print(f"Loaded {len(test_graphs)} test graph(s).")

    # E) Evaluate
    evaluate_edge_classifier(model, test_graphs, device=device)
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import HeteroConv, GCNConv
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, accuracy_score
import numpy as np
import optuna
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

######################
# 1) HeteroEdgeClassifier Model
######################
class HeteroEdgeClassifier(nn.Module):
    def __init__(self, hidden_channels, out_channels, num_classes=2):
        super().__init__()
        self.conv = HeteroConv({
            ('ip', 'network', 'ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
            ('ip', 'context', 'ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
            ('ip', 'knowledge', 'ip'): GCNConv(-1, hidden_channels, add_self_loops=False),
        }, aggr='sum')

        self.lin_node = nn.Sequential(
            nn.Linear(hidden_channels, hidden_channels * 2),
            nn.ReLU(),
            nn.Linear(hidden_channels * 2, out_channels),
            nn.ReLU()
        )

        self.lin_edge = nn.Sequential(
            nn.Linear(out_channels * 2, out_channels),
            nn.ReLU(),
            nn.Linear(out_channels, num_classes)
        )

    def forward(self, data):
        x_dict = self.conv(data.x_dict, data.edge_index_dict)
        x_dict['ip'] = self.lin_node(x_dict['ip'])
        return x_dict

    def edge_classify(self, z_src, z_dst):
        z_cat = torch.cat([z_src, z_dst], dim=-1)
        logits = self.lin_edge(z_cat)
        return logits

##############################
# 2) Focal Loss
##############################
def focal_loss(logits, targets, alpha=0.25, gamma=2.0):
    ce_loss = F.cross_entropy(logits, targets, reduction='none')
    pt = torch.exp(-ce_loss)
    return (alpha * (1 - pt)**gamma * ce_loss).mean()

##############################
# 3) Train Edge Classifier
##############################
def train_edge_classifier(model, graphs, epochs, device, alpha=0.25, gamma=2.0):
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-3)

    for epoch in range(1, epochs + 1):
        model.train()
        total_loss = 0.0

        for data in graphs:
            data = data.to(device)
            optimizer.zero_grad()

            x_dict = model(data)
            z_ip = x_dict['ip']

            loss_graph = 0.0
            for rel_type in data.edge_types:
                if 'edge_label' not in data[rel_type]:
                    continue
                labels = data[rel_type].edge_label
                if labels.numel() == 0:
                    continue

                edge_index = data[rel_type].edge_index
                src, dst = edge_index
                logits = model.edge_classify(z_ip[src], z_ip[dst])

                loss_rel = focal_loss(logits, labels, alpha=alpha, gamma=gamma)
                loss_graph += loss_rel

            loss_graph.backward()
            optimizer.step()
            total_loss += loss_graph.item()

    avg_loss = total_loss / len(graphs)
    return avg_loss

################################
# 4) Evaluate Edge Classifier
################################
def evaluate_edge_classifier(model, test_graphs, device):
    model.eval()
    model.to(device)

    all_scores = []
    all_labels = []

    for data in test_graphs:
        data = data.to(device)
        x_dict = model(data)
        z_ip = x_dict['ip']

        for rel_type in data.edge_types:
            if 'edge_label' not in data[rel_type]:
                continue
            labels = data[rel_type].edge_label
            if labels.numel() == 0:
                continue

            edge_index = data[rel_type].edge_index
            src, dst = edge_index
            logits = model.edge_classify(z_ip[src], z_ip[dst])

            probs = F.softmax(logits, dim=-1)[:, 1]
            all_scores.extend(probs.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())

    all_scores = np.array(all_scores)
    all_labels = np.array(all_labels)

    auc = roc_auc_score(all_labels, all_scores)
    preds = (all_scores > 0.5).astype(int)
    precision, recall, f1, _ = precision_recall_fscore_support(all_labels, preds, average='binary', zero_division=0)
    accuracy = accuracy_score(all_labels, preds)

    return auc, precision, recall, f1, accuracy, all_scores, all_labels

###############################
# 5) Optuna Optimization
###############################
def objective(trial):
    hidden_channels = trial.suggest_int("hidden_channels", 16, 64, step=16)
    out_channels = trial.suggest_int("out_channels", 8, 32, step=8)
    epochs = trial.suggest_int("epochs", 10, 100, step=10)

    model = HeteroEdgeClassifier(hidden_channels=hidden_channels, out_channels=out_channels)
    train_edge_classifier(model, train_graphs, epochs=epochs, device=device)

    auc, _, _, _, _, _, _ = evaluate_edge_classifier(model, test_graphs, device=device)

    return auc

###############################
# Main Script
###############################
if __name__ == "__main__":
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    train_dir = "3ed_trai_h_graphs_hetero_graphs"
    train_files = [f for f in os.listdir(train_dir) if f.endswith('.pt')]
    train_graphs = [torch.load(os.path.join(train_dir, f)) for f in train_files]

    test_dir = "3ed_tes_h_graphs_hetero_graphs"
    test_files = [f for f in os.listdir(test_dir) if f.endswith('.pt')]
    test_graphs = [torch.load(os.path.join(test_dir, f)) for f in test_files]

    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=500)

    best_trial = study.best_trial
    best_params = best_trial.params

    print(f"Best Hyperparameters: {best_params}")
    print(f"Best AUC: {best_trial.value}")

    # Train Final Model with Best Hyperparameters
    final_model = HeteroEdgeClassifier(hidden_channels=best_params['hidden_channels'], out_channels=best_params['out_channels'])
    train_edge_classifier(final_model, train_graphs, epochs=best_params['epochs'], device=device)

    # Save the Model
    torch.save(final_model.state_dict(), "best_model_3ed-2.pth")

    # Evaluate Final Model
    auc, precision, recall, f1, accuracy, all_scores, all_labels = evaluate_edge_classifier(final_model, test_graphs, device=device)

    print(f"Final Model Performance:\nAUC: {auc:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}, Accuracy: {accuracy:.4f}")

    # Plot AUC Curve
    fpr, tpr, _ = roc_curve(all_labels, all_scores)
    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC: {auc:.4f}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.savefig("best_model_auc_curve_3ed-3.png")
    plt.show()
```
